```python
import os
import io
import time
import pandas as pd
import pyperclip
from datetime import datetime

from PIL import Image
import win32clipboard

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException, NoSuchElementException, NoAlertPresentException

from webdriver_manager.chrome import ChromeDriverManager
from urllib.parse import urlparse
# Đọc file Excel

from openpyxl import load_workbook

# Đường dẫn đầy đủ đến file Excel
file_path = r'C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\Rawdata\CAPTURE\bot_log\bot_capture_log.xlsx'
sheet_name = 'Log'

# Kiểm tra file có tồn tại không
if not os.path.exists(file_path):
    print(f"❌ File không tồn tại: {file_path}")
else:
    try:
        # Mở workbook và sheet
        wb = load_workbook(file_path, data_only=True)
        ws = wb[sheet_name]

        # Đọc dữ liệu từ A1:G14
        data = []
        for row in ws['A1':'G14']:
            data.append([cell.value for cell in row])

        # Tạo DataFrame, lấy hàng đầu làm header
        df = pd.DataFrame(data[1:], columns=data[0])

        # Copy vào clipboard (có thể dán vào Excel)
        df.to_clipboard(index=False, excel=True)
        print("✅ Đã copy vùng A1:G14 vào clipboard.")
    except Exception as e:
        print(f"❌ Lỗi khi mở file hoặc đọc dữ liệu: {e}")
```
