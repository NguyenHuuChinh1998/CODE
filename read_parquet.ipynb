```python
import pandas as pd
from pathlib import Path # Nên dùng Path để quản lý đường dẫn

# 1. Xác định đường dẫn
# Tách riêng thư mục và tên file để dễ quản lý
base_folder = Path(r"C:\Users\huuchinh.nguyen\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\CODE\Resources")
file_name = "_rta_report_combination"

parquet_path = base_folder / f"{file_name}.parquet"
csv_path = base_folder / f"{file_name}.csv"

try:
    # 2. Đọc file parquet
    print(f"Đang đọc file: {parquet_path}")
    df = pd.read_parquet(parquet_path)
    print(f"Đã đọc xong {len(df)} dòng.")

    # 3. (MỚI) Ghi ra file CSV
    print(f"Đang ghi file CSV... (có thể mất vài giây): {csv_path}")
    
    # index=False: Bỏ cột index (0, 1, 2...) của pandas
    # encoding='utf-8-sig': Đảm bảo Excel đọc file CSV (đặc biệt là tiếng Việt) không bị lỗi font
    df.to_csv(csv_path, index=False, encoding='utf-8-sig')

    print("Hoàn thành! File đã được lưu.")

    # 4. Hiển thị 5 dòng đầu tiên (như code cũ của bạn)
    print("\n--- Thông tin 5 dòng đầu ---")
    print(df.columns)
    print(df.dtypes)
    display(df.head()) # Dùng display() sẽ đẹp hơn print(df) trong Jupyter

except FileNotFoundError:
    print(f"LỖI: Không tìm thấy file tại: {parquet_path}")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```
