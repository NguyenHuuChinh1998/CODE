```python
import polars as pl
import glob
import os

def run_diagnostic():
    # 1. Danh sách các file/thư mục cần quét
    folders_to_scan = [
        # Thư mục RAW CSV (chứa 'New Look Excel data_EN')
        r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\CSV\*New Look Excel data_EN*.csv",
        # Thư mục Output Combine (Quét cả CSV và Parquet)
        r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\OUTPUT_PERFORMANCE\OUTPUT_PERFORMANCE_COMBINE\*.csv",
        r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\OUTPUT_PERFORMANCE\OUTPUT_PERFORMANCE_COMBINE\*.parquet",
        # Thư mục Global Report
        r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\Global Report\performance_en\*.parquet"
    ]

    # 2. Định nghĩa mục tiêu
    target_emails = ["pranav.jadhav2@concentrix.com", "arbaz.shaikh4@concentrix.com"]
    correct_location = "Concentrix (Pune)"

    affected_files = []
    total_wrong_rows = 0

    print("🔍 BẮT ĐẦU CHẠY CHẨN ĐOÁN...\n" + "-"*50)

    # 3. Quét từng thư mục
    for pattern in folders_to_scan:
        files = glob.glob(pattern)
        for file_path in files:
            try:
                # Dùng scan (Lazy) để tiết kiệm RAM, chuyển sang read nếu file nhẹ
                if file_path.endswith('.csv'):
                    # Bỏ qua lỗi schema rác trong CSV bằng ignore_errors
                    df = pl.read_csv(file_path, ignore_errors=True, infer_schema_length=10000)
                else:
                    df = pl.read_parquet(file_path)

                # Kiểm tra xem file có chứa 2 cột cần thiết không
                if "Agent Email ID" in df.columns and "Agent Business Location" in df.columns:
                    
                    # Lọc ra các dòng có Email bị nhắm mục tiêu NHƯNG Location đang bị SAI
                    wrong_records = df.filter(
                        (pl.col("Agent Email ID").is_in(target_emails)) & 
                        (pl.col("Agent Business Location") != correct_location)
                    )

                    wrong_count = len(wrong_records)
                    
                    if wrong_count > 0:
                        print(f"⚠️ Phát hiện {wrong_count} dòng sai tại:")
                        print(f"   ➔ {file_path}")
                        affected_files.append(file_path)
                        total_wrong_rows += wrong_count

            except Exception as e:
                print(f"❌ Lỗi khi đọc file (có thể file hỏng hoặc khóa): {file_path}")
                print(f"   Chi tiết lỗi: {e}")

    # 4. Báo cáo tổng kết
    print("-" * 50)
    print("📊 KẾT QUẢ CHẨN ĐOÁN:")
    if len(affected_files) == 0:
        print("✅ Tuyệt vời! Không tìm thấy file nào bị sai Location cho 2 agent này.")
    else:
        print(f"🚨 Tổng số file bị ảnh hưởng: {len(affected_files)} files")
        print(f"🚨 Tổng số dòng cần sửa: {total_wrong_rows} dòng")
        print("\n💡 LỜI KHUYÊN: Hãy copy (backup) các file được liệt kê ở trên ra một thư mục riêng trước khi chạy code Replace.")

    return affected_files

# Kích hoạt hàm
list_of_files_to_fix = run_diagnostic()
```

```python
import os
import shutil
import polars as pl

# 1. Danh sách 6 file chính xác đã được phát hiện lỗi
files_to_fix = [
    r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\CSV\New Look Excel data_EN 2026-03.csv",
    r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\CSV\New Look Excel data_EN 2026-04.csv",
    r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\OUTPUT_PERFORMANCE\OUTPUT_PERFORMANCE_COMBINE\26_03.csv",
    r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\OUTPUT_PERFORMANCE\OUTPUT_PERFORMANCE_COMBINE\26_04.csv",
    r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\OUTPUT_PERFORMANCE\OUTPUT_PERFORMANCE_COMBINE\26_03_performance_en_vn.parquet",
    r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\RAW\OUTPUT_PERFORMANCE\OUTPUT_PERFORMANCE_COMBINE\26_04_performance_en_vn.parquet"
]

target_emails = ["pranav.jadhav2@concentrix.com", "arbaz.shaikh4@concentrix.com"]
correct_location = "Concentrix (Pune)"

# 2. Tạo đường dẫn thư mục làm việc mới (An toàn tuyệt đối)
base_dir = r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\BI_Task\FIX_LOCATION_TASK"
backup_dir = os.path.join(base_dir, "BACKUP_ORIGINAL")
fixed_dir = os.path.join(base_dir, "FIXED_READY_TO_COPY")

os.makedirs(backup_dir, exist_ok=True)
os.makedirs(fixed_dir, exist_ok=True)

print("🚀 BẮT ĐẦU QUÁ TRÌNH BACKUP VÀ SỬA LỖI...\n" + "-"*50)

for file_path in files_to_fix:
    filename = os.path.basename(file_path)
    
    try:
        # BƯỚC A: COPY FILE GỐC SANG THƯ MỤC BACKUP
        backup_path = os.path.join(backup_dir, filename)
        shutil.copy2(file_path, backup_path)
        print(f"✅ Đã Backup an toàn: {filename}")
        
        # BƯỚC B: ĐỌC DỮ LIỆU TỪ FILE BACKUP (Tránh lỗi khóa file gốc)
        if filename.endswith(".csv"):
            df = pl.read_csv(backup_path, ignore_errors=True, infer_schema_length=10000)
        else:
            df = pl.read_parquet(backup_path)
            
        # BƯỚC C: REPLACE LOCATION THÀNH PUNE CHO 2 EMAIL
        df_fixed = df.with_columns(
            pl.when(pl.col("Agent Email ID").is_in(target_emails))
            .then(pl.lit(correct_location))
            .otherwise(pl.col("Agent Business Location"))
            .alias("Agent Business Location")
        )
        
        # BƯỚC D: XUẤT FILE ĐÃ SỬA RA THƯ MỤC FIXED
        fixed_path = os.path.join(fixed_dir, filename)
        if filename.endswith(".csv"):
            df_fixed.write_csv(fixed_path)
        else:
            df_fixed.write_parquet(fixed_path)
            
        print(f"✨ Đã Replace thành công và lưu tại thư mục FIXED: {filename}\n")
        
    except Exception as e:
        print(f"❌ Xảy ra lỗi với file {filename}: {e}\n")

print("-" * 50)
print("🎉 HOÀN TẤT CHIẾN DỊCH!")
print(f"📂 Nơi chứa file dự phòng (phòng hờ): {backup_dir}")
print(f"📂 Nơi chứa các file ĐÃ SẠCH LỖI: {fixed_dir}")
print("👉 Bây giờ bạn có thể mở thư mục FIXED_READY_TO_COPY và copy đè thủ công vào các thư mục gốc.")
```
