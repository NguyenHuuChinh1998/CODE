```python
import pandas as pd
import datetime
```

```python
monday_start = 'SUN'
tuesday_start = 'MON'
wednesday_start = 'TUE'
thursday_start = 'WED'
friday_start = 'THU'
saturday_start = 'FRI'
sunday_start = 'SAT'

monday_end = 'MON'
tuesday_end = 'TUE'
wednesday_end = 'WED'
thursday_end = 'THU'
friday_end = 'FRI'
saturday_end = 'SAT'
sunday_end = 'SUN'

def create_weekly_dataframe(column_name, start_date, end_date, start_day, end_day):
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')
    df = pd.DataFrame({'Mod_date': date_range})
    start_day = start_day.upper()
    end_day = end_day.upper()
    start_time = df['Mod_date'].dt.to_period('W-' + start_day).dt.start_time
    end_time = df['Mod_date'].dt.to_period('W-' + end_day).dt.end_time
    df[column_name] = 'W' + df['Mod_date'].dt.strftime('%U').str.lstrip('0') + ' (' + start_time.dt.strftime('%m.%d') + ' - ' + end_time.dt.strftime('%m.%d') + ')'
    return df,start_day,end_day
```

**CIRCLE 1 (SAT - THU) - [mở dự án -> 14.04.2023]**

```python
start_date_circle_1 = datetime.datetime(2022, 12, 26)
end_date_circle_1 =  datetime.datetime(2023, 4, 14)

df_regular_circle_1,start_day_regular,end_day_regular = create_weekly_dataframe('Regular_Week', start_date_circle_1, end_date_circle_1, saturday_start, friday_end)
df_upl_circle_1,start_day_upl,end_day_upl = create_weekly_dataframe('UPL_Week', start_date_circle_1, end_date_circle_1, monday_start, sunday_end)
df_labeling_circle_1,start_day_label,end_day_label = create_weekly_dataframe('Labeling_Week', start_date_circle_1, end_date_circle_1, thursday_start, wednesday_start)
df_bluearmy_circle_1,start_day_label,end_day_bluearmy = create_weekly_dataframe('Blue_Army_Week', start_date_circle_1, end_date_circle_1, thursday_start, wednesday_start)
df_report_circle_1,start_day_report,end_day_report = create_weekly_dataframe('Report_Week', start_date_circle_1, end_date_circle_1, monday_start, sunday_end)

merged_df_circle_1 = [df_regular_circle_1, df_upl_circle_1, df_labeling_circle_1,df_bluearmy_circle_1,df_report_circle_1]

df_timeline_expand_circle_1 = merged_df_circle_1[0]
for df in merged_df_circle_1[1:]:
    df_timeline_expand_circle_1 = pd.merge(df_timeline_expand_circle_1, df, on='Mod_date')

df_timeline_expand_circle_1.insert(1, 'Days', pd.to_datetime(df_timeline_expand_circle_1['Mod_date']).dt.strftime('%a').str.upper())

for index, row in df_timeline_expand_circle_1.iterrows():
    if index < len(df_timeline_expand_circle_1) - 1 and row['Days'] == end_day_regular:
        df_timeline_expand_circle_1.at[index, 'Regular_Week'] = df_timeline_expand_circle_1.at[index, 'Regular_Week']
```
