```python
def convert_datetime(a):
    if pd.isnull(a):
        return pd.NaT
    
    date_formats = ["%Y-%m-%d %H:%M:%S.%f", 
                    "%Y-%m-%d %H:%M:%S", 
                    "%m/%d/%Y %H:%M",
                    "%m-%d-%Y %H:%M", 
                    "%m/%d/%Y", 
                    "%Y-%m-%d", 
                    "%H:%M.%f",
                    "%m/%d/%Y %H:%M:%S"] 
    for fmt in date_formats:
        try:
            return datetime.strptime(str(a), fmt).strftime("%Y-%m-%d" )
        except ValueError:
            pass
    try:
        return (datetime(1899, 12, 30) + timedelta(days=float(a))).strftime("%Y-%m-%d")
    except ValueError:
        pass
    if not re.match("\d+:\d+:\d+(\.\d+)?", str(a)):
        return str(a)

def convert_time(a):
    if pd.isnull(a):
        return pd.NaT
    
    date_formats = ["%Y-%m-%d %H:%M:%S.%f", 
                    "%Y-%m-%d %H:%M:%S", 
                    "%m/%d/%Y %H:%M",
                    "%m-%d-%Y %H:%M", 
                    "%m/%d/%Y", 
                    "%Y-%m-%d", 
                    "%H:%M.%f",
                    "%m/%d/%Y %H:%M:%S"] 
    for fmt in date_formats:
        try:
            return datetime.strptime(str(a), fmt).strftime("%Y-%m-%d %H:%M:%S")
        except ValueError:
            pass
    try:
        return (datetime(1899, 12, 30) + timedelta(days=float(a))).strftime("%Y-%m-%d %H:%M:%S")
    except ValueError:
        pass
    if not re.match("\d+:\d+:\d+(\.\d+)?", str(a)):
        return str(a)
```

```python
def merge_csv_files(folder_path):
    file_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

    dfs = []
    for file_path in file_paths:
        try:
            df = pd.read_csv(file_path)
            df['source_file'] = os.path.basename(file_path)

            dfs.append(df)
        except Exception as e:
            print(f"Error reading file {file_path}: {str(e)}")
  
    merged_df = pd.concat(dfs, axis=0, ignore_index=True)

    return merged_df
```

```python
def concat_files(folder_path):
    files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx')]
    
    if len(files) == 0:
        print("Thư mục không chứa file xlsx.")
        return None
    
    first_file_path = os.path.join(folder_path, files[0])
    df_header = pd.read_excel(first_file_path, header=0)
    
    num_columns = df_header.shape[1]
    
    dfs = []
    for file in files[1:]:
        file_path = os.path.join(folder_path, file)
        df = pd.read_excel(file_path, header=0)
        if df.shape[1] == num_columns:
            dfs.append(df)
    df_combined = pd.concat([df_header] + dfs)
    
    return df_combined

mod_interval_hour_raw = concat_files('D:/Shrinkage/MOD Interval (hour)/xlsx')
```
