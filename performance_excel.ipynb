```python
import win32com.client
import subprocess
import time
import os

FILE_PATH = r"C:\Users\ADMIN\Concentrix Corporation\WFM-Expedia-HCM - Branding files\Performance\Performance_Report_SendOut\Retail_Performance_Expedia_VN _Update_Apr_2026.xlsb"

def pause_onedrive():
    subprocess.Popen(["C:\\Program Files\\Microsoft OneDrive\\OneDrive.exe", "/pause"])
    time.sleep(3)

def resume_onedrive():
    subprocess.Popen(["C:\\Program Files\\Microsoft OneDrive\\OneDrive.exe"])

def refresh_all(file_path: str, save: bool = True):
    pause_onedrive()
    try:
        excel = win32com.client.Dispatch("Excel.Application")
        excel.Visible = False
        excel.DisplayAlerts = False

        wb = excel.Workbooks.Open(os.path.abspath(file_path))
        wb.RefreshAll()
        excel.CalculateUntilAsyncQueriesDone()

        if save:
            wb.Save()

        wb.Close()
        excel.Quit()
    finally:
        resume_onedrive()

refresh_all(FILE_PATH)
```
